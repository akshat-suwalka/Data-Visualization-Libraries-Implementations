## Library

```
import numpy as np
```

## Data

```
x = np.arange(0,100)
y = x*2
z = x**2
```

```
import matplotlib.pyplot as plt
%matplotlib inline
```

## Plot a st. line graph by its two coordinates

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
ax.plot(x,y)
```

## SubPlotting

```
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y,color = 'red')
ax2.plot(x,y,color = 'blue')
```

## Subplotting 2

```
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,z)
ax2.plot(x,y)

ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

## Subplot 3 

```
fig,axes = plt.subplots(1,2)

axes[0].plot(x,y,color = 'red', ls = '--')
axes[1].plot(x,z,color = 'blue', lw = 3)
```

```
fig,axes = plt.subplots(1,2,figsize = (12,2))

axes[0].plot(x,y,color = 'red', ls = '--')
axes[1].plot(x,z,color = 'blue', lw = 3)
```
